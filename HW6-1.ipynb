```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sympy import diff, Symbol
```

```python
X = np.linspace(4,13,10000)
Y = (X - 4)*(X - 5.5)*(X - 9)*(X - 10.8)*(X - 13)
plt.scatter(X,Y)
plt.show
def y(x):
    return (x - 4)*(x - 5.5)*(x - 9)*(x - 10.8)*(x - 13)
```

```python
import random
a = 6
h = 0.0001
k = 0.01
A = []
while(1):
    m = (y(a+h)-y(a-h))/(2*h)
    print(a,m)
    A.append(a)
    if(m > 0.001):
        a = a - k * m
    elif(m < -0.001):
        a = a - k * m
    else:
        break
```

```python
A = np.array(A)
X = np.linspace(5,9,10000)
Y = (X - 4)*(X - 5.5)*(X - 9)*(X - 10.8)*(X - 13)
plt.plot(X,Y)
Y1 = (A - 4)*(A - 5.5)*(A - 9)*(A - 10.8)*(A - 13)
plt.plot(A,Y1,'r')
```
