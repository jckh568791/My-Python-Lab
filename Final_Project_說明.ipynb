主題：辨識居家常見蟲類

介紹：鑑於許多人在家中看到不知名的蟲，經常會在社群網站上詢問大眾，藉此得知如何應付此昆蟲，這通常需要等待時間才能獲得答案，因此希望能夠建立一個模型，可以直接拍照就得知此昆蟲的資料

實作方法：

1.上網找尋居家常見昆蟲圖片並標記

   a.上網搜尋圖片

   b.轉換圖片為統一格式(128x128, 灰階)

   c.完成x_train, x_test, y_train, y_test 之array

2.透過CNN訓練，由照片辨識此昆蟲名稱，並提供此昆蟲之相關資料以及應對方式（Ex:益蟲or害蟲、是否有毒...）

專案排程：

5/24 初步規劃，完成批次處理圖片程式，並開始收集訓練資料(上網下載圖片)

5/31 完成資料收集、標記

6/03 完成訓練CNN模型

6/09 上傳期末報告

成果報告：

經過900次訓練，正確率達到62.79%
