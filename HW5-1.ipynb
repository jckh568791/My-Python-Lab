```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
```

```python
df = pd.read_html("http://stat.motc.gov.tw/mocdb/stmain.jsp?sys=220&ym=6000&ymt=10700&kind=21&type=1&funid=b210101&cycle=4&outmode=0&compmode=0&outkind=1&fldspc=0,7,&rdm=namNIric")
df_1 = df[1]
df_1.head()
```

```python
import seaborn as sns
sns.set_style('whitegrid') 
from matplotlib.font_manager import FontProperties
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
year = []
for i in range(1, len(df_1) - 1): #僅保留數字部分
    a = list(filter(str.isdigit, df_1[0][i]))
    b = int(''.join(str(i) for i in a))
    year.append(b)
trip = []
for i in range(1, len(df_1) - 1):
    trip.append(int(df_1[1][i])/10000)
plt.scatter(year, trip)
plt.xlabel("年")
plt.ylabel("旅客人數(萬人次)")
plt.ticklabel_format(style='plain', axis='y')
plt.title("臺灣鐵路管理局旅客列車客運概況")
plt.show
```

```python
from sklearn.linear_model import LinearRegression
from pandas.core.frame import DataFrame
regr = LinearRegression()
regr.fit(year,trip)
Trip = regr.predict(year)
plt.scatter(year, trip)
plt.plot(year, Trip, 'r')
plt.xlabel("年")
plt.ylabel("旅客人數(萬人次)")
plt.ticklabel_format(style='plain', axis='y')
plt.title("臺灣鐵路管理局旅客列車客運概況")
plt.show
```
